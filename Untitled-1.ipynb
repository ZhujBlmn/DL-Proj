```python
import json
import numpy as np
import matplotlib.pyplot as plt

# ---  ---
TRAIN_LOSS_FILE = 'models/controlnet_video_frame/training_loss.json'
VAL_LOSS_FILE = 'models/controlnet_video_frame/validation_loss.json' 

SMOOTHING_WINDOW = 100  #  Loss 

def load_loss_data(file_path):
    """ JSON """
    try:
        with open(file_path, 'r') as f:
            loss_list = json.load(f)
        return np.array(loss_list)
    except FileNotFoundError:
        print(f" '{file_path}'")
        return None

def smooth_loss(loss_array, window_size):
    """ Loss """
    kernel = np.ones(window_size) / window_size
    smoothed = np.convolve(loss_array, kernel, mode='valid')
    # 
    padding = np.ones(window_size - 1) * smoothed[0]
    return np.concatenate((padding, smoothed))

def plot_loss_from_array(train_losses, val_losses):
    if train_losses is None: return
    
    total_steps = len(train_losses)
    train_x = np.arange(total_steps)

    # 1.  Loss 
    smoothed_train = smooth_loss(train_losses, SMOOTHING_WINDOW)
    
    # 2.  Loss  Step 
    #  Loss  Epoch  Epoch  Step
    if val_losses is not None:
        num_epochs = len(val_losses)
        steps_per_epoch = total_steps // num_epochs
        #  X [steps_per_epoch, 2*steps_per_epoch, ..., total_steps]
        val_x = np.arange(1, num_epochs + 1) * steps_per_epoch
    
    # 3. 
    plt.figure(figsize=(12, 6), dpi=100)
    
    # 4. 
    plt.plot(train_x, train_losses, 
             label='Original Training Loss (Step)', 
             color='lightsteelblue', alpha=0.4, linewidth=0.5)
    
    # 5. 
    valid_start = SMOOTHING_WINDOW - 1
    plt.plot(train_x[valid_start:], smoothed_train[valid_start:], 
             label=f'Smoothed Training Loss (Window={SMOOTHING_WINDOW})', 
             color='red', linewidth=1)

    # 6. 
    if val_losses is not None:
        plt.plot(val_x, val_losses, 
                 label='Validation Loss (per Epoch)', 
                 color='blue', marker='o', markersize=4, linewidth=1.5)
    
    # 7. 
    plt.title('ControlNet Training & Validation Loss (SSV2)')
    plt.xlabel(f'Training Steps (Total: {total_steps:,})')
    plt.ylabel('MSE Loss (Noise Prediction)')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
    
    #  Y  Loss 
    plt.ylim(bottom=0.0, top=min(1.1, np.max(train_losses)*1.1))
    
    plt.legend(loc='upper right')
    plt.tight_layout()
    
    # 
    plt.savefig('training_validation_curve.png')
    plt.show()

if __name__ == '__main__':
    t_loss = load_loss_data(TRAIN_LOSS_FILE)
    v_loss = load_loss_data(VAL_LOSS_FILE)
    plot_loss_from_array(t_loss, v_loss)
```
